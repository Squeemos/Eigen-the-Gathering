```python
import os
import datetime

import numpy as np
import pandas as pd
```

## Load Data

```python
path = "data/"

data_files = [fn for fn in os.listdir(path) if "json" in fn]
data_filename = path + data_files[-1]
data_filename
```

```python
with open(data_filename, "r") as infile:
    data = pd.read_json(infile)
data
```

## Wranglin'

```python
for i, col in enumerate(data.columns):
    print(col, end=", ")
    if (i + 1) % 5 == 0:
        print()
```

```python
# Add timestamp
data["utc"] = pd.to_datetime("today").now()
```

```python
def list_to_str(data: pd.DataFrame, column_name: str):
    return data[column_name].apply(
        lambda x: ",".join(x) if isinstance(x, list) else None
    )

list_to_str(data, "promo_types")
list_to_str(data, "frame_effects")
```

```python
dfs = {
    "cards": data.loc[:, ["id", "name", "set_name", "border_color", "promo_types", "frame_effects"]],
    "prices": pd.concat([data.loc[:, ["id", "utc"]], pd.json_normalize(data.prices)], axis=1),
    "images": pd.concat([data.loc[:, ["id", "image_status"]], pd.json_normalize(data.image_uris)], axis=1),
}
```

```python
dfs["cards"]["promo_types"] = list_to_str(dfs["cards"], "promo_types")
dfs["cards"]["frame_effects"] = list_to_str(dfs["cards"], "frame_effects")
```

## Database

```python
import sqlite3
```

```python
# Open database
db_path = os.path.normpath(os.getcwd() + "/data/demo.db")
conn = sqlite3.connect(db_path)
cursor = conn.cursor()
```

```python
# Insert data
dfs["cards"].to_sql("Cards", conn, if_exists="replace", index=False)
dfs["prices"].to_sql("Prices", conn, if_exists="append", index=False)
dfs["images"].to_sql("Images", conn, if_exists="replace", index=False)
```

```python
# Show heads of tables
for table in ("Cards", "Prices", "Images"):
    cursor.execute(f"SELECT * FROM {table}")

    print(f"=== {table} ===")

    col_names = [col_desc[0] for col_desc in cursor.description]
    print(col_names)

    rows = cursor.fetchall()
    for row in rows[:5]:
        print(row)

    print()
```

```python
# Close connection
conn.close()
```
