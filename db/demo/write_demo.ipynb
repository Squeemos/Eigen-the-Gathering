```python
import os
import datetime

import pandas as pd
```

## Load Data

```python
path = "data/"

data_files = [fn for fn in os.listdir(path) if "json" in fn]
data_filename = path + data_files[-1]
data_filename
```

```python
with open(data_filename, "r") as infile:
    data = pd.read_json(infile)
data
```

## Wranglin'

```python
for i, col in enumerate(data.columns):
    print(col, end=", ")
    if (i + 1) % 5 == 0:
        print()
```

```python
# Add timestamp
data["utc"] = pd.to_datetime("today").now()
```

```python
pd.json_normalize(data.image_uris)
```

```python
dfs = {
    "cards": data.loc[:, ["id", "name", "set_name"]],
    "prices": pd.concat([data.loc[:, ["id", "utc"]], pd.json_normalize(data.prices)], axis=1),
    "images": pd.concat([data.loc[:, ["id", "image_status"]], pd.json_normalize(data.image_uris)], axis=1),
}
```
